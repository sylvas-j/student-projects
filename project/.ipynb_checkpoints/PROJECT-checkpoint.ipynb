#### SCRAPE ANY WEBSITE AND GET A USEFUL OUTPUT

#### SCRAPING THE OFFICE OF THE SECRETARY TO THE GOVERTMENT OF THE FEDERATION (OSGF) WEBSITE TO GET THE NAMES OF ALL THE  PRESIDENTS OF NIGERIA SINCE AFTER INDEPENDENCE

```python
from bs4 import BeautifulSoup as bs
import urllib
url = "https://www.osgf.gov.ng/"
page = urllib.request.urlopen(url)
soup = bs(page, 'html.parser')
soup
```

#### TAG AND CLASS

```python
president = soup.find_all('a', class_="dropdown-item")
print(president)
```

#### NUMBER OF ITEMS

```python
list_of_items = soup.find_all('a', class_="dropdown-item")
```

```python
len(list_of_items)
```

```python
for item in list_of_items:
    # print(items.get_text().replace("\n",''))
    item= item.get_text().replace('\n','')
    print(item.strip(' '))
```

#### GET TABLE

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.osgf.gov.ng/about-us/history/past-presidents-and-heads-state'

# get the webpage
response = requests.get(url)

# create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# find the table
table = soup.find('table')

# iterate through the table rows 
for tr in table.find_all('tr'):
    print('\n')
    # iterate through the table columns
    for td in tr.find_all('td'):
        print(f'{td.text:15}', end='\t')
```
